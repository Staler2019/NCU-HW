## 第一題
- 獲取數組a和b之間的共同元素

```
#1
import numpy as np
a = np.array([1,2,3,2,3,4,3,4,5,6])
b = np.array([7,2,10,2,7,4,9,4,9,8])
answer1 = np.intersect1d(a,b)
print(answer1)
```

## 第四題
- 將max函數導入到a,b數列使得它能夠取出每行最大的出來組成一個新的數列answer4

```
#4
def max(x,y):
  if x >= y:
    return x
  else:
    return y

a = np.array([5,7,9,8,6,4,5])
b = np.array([6,3,4,8,9,7,1])

pair_max = np.vectorize(max)
answer4 = pair_max(a,b)
print(answer4)
```

## 第五題
- 創建一個標準化形式的iris的sepallength，其值的範圍在0和1之間，最小值為0，最大值為1。
  - 公式:((數列中每個元素)-(數列中最小的元素)) / ((數列中最大的元素) - (數列中最小的元素))

```
#5

sepallength = np.genfromtxt(url, dtype="float",delimiter=',',usecols=[0])
maxFloat = sepallength.max()
minFloat = sepallength.min()

answer5 = (sepallength-minFloat)/(maxFloat-minFloat)
print(answer5)
```

## 第六題
- 計算sepallength的softmax分數。
- image.png

- 這邊的 ${y_i}$選擇成每個值減去數列中最大的元素

```
#6
sepallength = np.genfromtxt(url, dtype="float", delimiter=',', usecols=[0])

def softmax(x):
  e_x = np.exp(x - np.max(x))
  return e_x / e_x.sum()

answer6 = softmax(sepallength)
print("answer6:",answer6)
```
