```
#2

# [-1,k+1,-5,-1]
# 1 + k**2 + 2*k + 1 + 25 + 1 = 36
# k**2 + 2*k + 8 = 0
def quadratic(a,b,c):
    key= b**2 - 4*a*c #判別式
    if key>0:
        x1=(-b + key**(1/2)) / (2*a)
        x2=(-b - key**(1/2)) / (2*a)
    if key==0:
        x1=(-b) / (2*a)
        x2=(-b) / (2*a)
    if key<0:
        print('方程無解')
        return(None,None)
    return (x1,x2)
print(quadratic(1,2,-8)) #代入你要解的方程式係數
```
